```python
x=data.drop("Subscription Status",axis=1)
y=data["Subscription Status"]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
categorical_features=["Gender","Item Purchased","Category","Location","Size","Color","Season","Payment Method","Shipping Type"]
```

```python
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,categorical_features)],remainder="passthrough")
transformed_x=transformer.fit_transform(x)
pd.DataFrame(transformed_x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
np.random.seed(42)
x_train,x_test,y_train,y_test=train_test_split(transformed_x,y,test_size=0.2)
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(model,transformed_x,y,cv=10)
score
```

```python
final_score=score.mean()
print(f"The final score of the logistic regression is {final_score*100:.2f} %")
```

```python
y_preds=model.predict(x_test)
y_preds[0:30]
```

```python
np.array(y_test)[0:30]
```

```python
from sklearn.ensemble import RandomForestClassifier
new_model=RandomForestClassifier()
new_model.fit(x_train,y_train)
new_model.score(x_test,y_test)
```

```python
score_new=cross_val_score(new_model,transformed_x,y,cv=10)
score_new
```

```python
final_score_classifier=score_new.mean()
print(f"The random forest Classifier score is {final_score_classifier*100:.2f} %")
```
